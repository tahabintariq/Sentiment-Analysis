```python
positive_reviews_sentiments = []
negative_reviews_sentiments = []

def calculate_sentiment_score(review, words_dict):
    sentiment_score = 0
    for word in review.split():
        if word in words_dict:
            sentiment_score += words_dict[word]
    return sentiment_score

for review in positive_list:
    sentiment_score = calculate_sentiment_score(review, positive_words)
    positive_reviews_sentiments.append(sentiment_score)

for review in negative_list:
    sentiment_score = calculate_sentiment_score(review, negative_words)
    negative_reviews_sentiments.append(sentiment_score)

print("Sentiment scores for positive reviews:", positive_reviews_sentiments)
print("Sentiment scores for negative reviews:", negative_reviews_sentiments)
```

# Storage

```python
combined_reviews = positive_list + negative_list
combined_sentiment_scores = positive_reviews_sentiments + negative_reviews_sentiments

sentiment_labels = {}

for i, review in enumerate(combined_reviews):
    sentiment_score = combined_sentiment_scores[i]
    if sentiment_score > 1:
        sentiment_labels[review] = "positive"
    elif sentiment_score < 1:
        sentiment_labels[review] = "negative"
    elif 0.9 <= sentiment_score <= 1.1:
        sentiment_labels[review] = "neutral"

print(dict(list(sentiment_labels.items())[:30]))

```

```python
file_name = "sentiment_labels.txt"

with open(file_name, 'w') as file:
    for review, label in sentiment_labels.items():
        file.write(f"{review}: {label}\n")
```
